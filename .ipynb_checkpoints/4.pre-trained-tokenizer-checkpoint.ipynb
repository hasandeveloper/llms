```python
# Here we use pretrained tokenizers module which we were doing this manual in NLP and pipeline works under the hood which we used in last in 3.hugging-face-transformers.ipynb also we have different types of tokenizers which different module needs it own types
from transformers import AutoTokenizer
# BAsically some module no need of tokenization and some module needs
```

```python
model = "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
sentence = "I'm so excited to be learning about large language models"
```

```python
input_ids = tokenizer(sentence)
print(input_ids)
```

```python
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
print(token_ids)
```

```python
decoded_ids = tokenizer.decode(token_ids)
print(decoded_ids)
```

```python
tokenizer.decode(102)
```

```python
tokenizer.decode(101)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
print(sentence)
print(input_ids)
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
input_ids_pt = tokenizer(sentence, return_tensors ="pt")
print(input_ids_pt)
```

```python
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
with torch.no_grad():
    logits = model(**input_ids_pt).logits

predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
```
